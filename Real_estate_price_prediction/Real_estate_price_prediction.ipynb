```python
import pandas as pd #use to make dataframe
import numpy as np #use to perform matrix operation
from matplotlib import pyplot as plt #use to make plot and graph
%matplotlib inline #To appear your graph and plots into the notebook itself not in the pop up window
import matplotlib #use to make plot and graph
matplotlib.rcParams["figure.figsize"] = (20,10) #defining the size of the output window
```

```python
data_f1 = pd.read_csv("bengaluru_house_data.csv")
data_f1.head(10)
```

```python
data_f1.columns #showing all columns
```

```python
data_f2 = data_f1.drop(['area_type','society','balcony','availability'],axis='columns') #dropping all columns
#and saving it into new dataframe
data_f2.head(10) #showing 10 rows of data
```

```python
data_f2.isnull().sum() #Number of Nan Values
```

```python
data_f3 = data_f2.dropna()
data_f3.isnull().sum()
```

```python
data_f3['bhk'] = data_f3['size'].apply(lambda x: int(x.split(' ')[0]))
data_f3.head(10)
```

```python
data_f3.bhk.unique()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
data_f3[~data_f3['total_sqft'].apply(is_float)].head(10)
```

```python
def convert(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return ((float(tokens[0]) + float(tokens[1]))/2)
    try:
        return float(x)
    except:
        return None
```

```python
data_f4 = data_f3.copy()
data_f4.total_sqft = data_f4.total_sqft.apply(convert)
data_f4 = data_f4[data_f4.total_sqft.notnull()]
data_f4.head(10)
```

```python
data_f5 = data_f4.copy()
data_f5['price_rate'] = data_f5['price']*100000/data_f5['total_sqft']
data_f5.head()
```

```python
data_f5.location = data_f5.location.apply(lambda x: x.strip())
location_count = data_f5.groupby('location')['location'].agg('count').sort_values(ascending = False)
#dataframe.groupby() function is used to split the data into groups based on some criteria.
location_count.head(20)
```

```python
lesser_location_count = location_count[location_count<=10]
lesser_location_count
```

```python
data_f5.location = data_f5.location.apply(lambda x: 'other' if x in lesser_location_count else x)
#data_f5.head(10)
len(data_f5.location.unique())
```

```python
data_f5[data_f5.total_sqft/data_f5.bhk<300].head()
```

```python
data_f6 = data_f5[~(data_f5.total_sqft/data_f5.bhk<300)]
data_f6.shape
```

```python
data_f6.price_rate.describe()
```

```python
#Outliers: These are some observations which lies outside our expectation... at the abnormal distance from others.
#Standard deviation is the excellent way to find the outlier.
```

```python
#In this function if the distnce of the observation is more than one standard devition from means, then that
#point will not be considered hence that point is a outlier.
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_rate)
        st = np.std(subdf.price_rate)
        reduced_df = subdf[(subdf.price_rate>(m-st)) & (subdf.price_rate<=(m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
#new dataframe after reomving outliers.
data_f7 = remove_pps_outliers(data_f6)
data_f7.shape
```

```python
data_f7[data_f7.bath>data_f7.bhk+2]
```

```python
data_f8 = data_f7[data_f7.bath<data_f7.bhk+2]
data_f8.shape
```

```python
data_f9 = data_f8.drop(['size','price_rate'],axis='columns')
data_f9.head()
```

```python
dummies = pd.get_dummies(data_f9.location)
dummies.head(3)
```

```python
data_f10 = pd.concat([data_f9,dummies.drop('other',axis='columns')],axis='columns')
data_f10.head()
```

```python
data_f11 = data_f10.drop('location',axis='columns')
data_f11.head(2)
```

```python
X = data_f11.drop(['price'],axis='columns')
X.head(3)
```

```python
y = y = data_f11.price
y.head(3)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train,y_train)
lr_clf.score(X_test,y_test)
```

```python
def predict_price(location,sqft,bath,bhk):    
    loc_index = np.where(X.columns==location)[0][0]
    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if loc_index >= 0:
        x[loc_index] = 1
    return lr_clf.predict([x])[0]
```

```python
predict_price('1st Phase JP Nagar',1000, 2, 2)
```

```python
predict_price('Indira Nagar',1000, 2, 2)
```
